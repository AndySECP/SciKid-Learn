```python
from sklearn.feature_extraction.text import CountVectorizer
question = data0['final_cleaned']
vect = CountVectorizer()
vect.fit(question)
bag = vect.transform(question)
vect_df = pd.DataFrame(bag.toarray(), columns=vect.get_feature_names())
```

```python
n0 ,c0 = data0.shape
```

```python
from sklearn.cluster import KMeans
q, w = vect_df.shape
vect_question = bag.toarray()
k = 4 # number of clusters
kmeans = KMeans(n_clusters=k, init='random')
kmeans.fit(vect_question)
#
# Get a list of the questions in each cluster
#
clusters = []
for j in range(0,k):
    clusters.append([])
for i in range(0,n0):
    clusters[kmeans.labels_[i]].append(i)
#
# Print out clusters
#
for j in range(0,k):
    print(j+1,'\n', clusters[j], '\n\n')
```

```python
cluster1 = [0 for i in range(n)]
data['cluster1'] = cluster1
for i in range(n0):
    data.loc[data0['index'][i], 'cluster1'] = kmeans.labels_[i]+5
```
