```python
import nltk
```

```python
nltk.download("punkt")
```

```python
question_words = nltk.word_tokenize(Elementary1['question'][0])
question_words
```

```python
nltk.download("stopwords")
import string
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation)

def build_bag_of_words(words):
    return {
        word:1 for word in words \
        if not word in useless_words}
```

```python
Biology = Elementary1['question'].str.replace(r"\(([A-Za-z0-9 _]+)\)", "")
```

```python
Biology.head()
```

```python
len(Biology)
```

```python
Biology_words = [0 for i in range(len(Biology))]
for i in range (len(Biology)):
    Biology_words[i] = (build_bag_of_words(nltk.word_tokenize(Biology[i])), 'Biology')
Biology_words[0]
```

```python
Physics = Elementary2['question'].str.replace(r"\(([A-Za-z0-9 _]+)\)", "")
```

```python
len(Physics)
```

We need to use the same number of question in Biology and in Physics.

```python
Physics_words = [0 for i in range (75)]
for i in range (75):
    Physics_words[i] = ((build_bag_of_words(nltk.word_tokenize(Physics[i]))), 'Physics')
Physics_words[1]
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(Biology_words+Physics_words)
```

```python
classifier.show_most_informative_features()
```

```python
Unknown = Elementary3['question'].str.replace(r"\(([A-Za-z0-9 _]+)\)", "")
Unknown_words = [0 for i in range(len(Unknown))]
for i in range (len(Unknown)):
    Unknown_words[i] = build_bag_of_words(nltk.word_tokenize(Unknown[i]))
Unknown_words[0]
```

```python
nltk.classify.util.accuracy(classifier, Biology_words+Physics_words)*100
```

```python
subject3 = []
for element in Unknown_words:
    subject3.append(classifier.classify(element))
```

```python
subject3
```

```python
Elementary3['subject']=subject3
```
